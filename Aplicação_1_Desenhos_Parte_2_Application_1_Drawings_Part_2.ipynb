```python
import cv2
import matplotlib.pyplot as plt

figure = cv2.imread("Write the image filename here")

figure = cv2.cvtColor(figure, cv2.COLOR_BGR2RGB)

plt.imshow(figure)
plt.show()
```

```python
# Primeira modificação efetuada na segunda imagem
# First modification made to the second image

# Mesmos passos da primeira modificação realizada na primeira imagem
# Same steps as the first modification performed on the first image.

figure_cp = figure.copy()
figure_cp[:,:,0] = figure[:,:,1]
figure_cp[:,:,1] = figure[:,:,2]
figure_cp[:,:,2] = figure[:,:,2]

fig,axs = plt.subplots(1, 2, layout='constrained', figsize = [13,13])

axs[0].imshow(figure)

axs[0].set_title('Imagem original / Original Image')

axs[0].set_axis_off()

axs[1].imshow(figure_cp)

axs[1].set_title('Imagem modificada / Modified Image')

axs[1].set_axis_off()

plt.show()
```

```python
# Segunda modificação efetuada na segunda imagem
# Second modification made to the second image

# Mesmos passos da segunda modificação realizada na primeira imagem
# Same steps as the second modification performed on the first image.

figure_cp = figure.copy()
figure_cp[:,:,0] = figure[:,:,0]
figure_cp[:,:,1] = figure[:,:,2]
figure_cp[:,:,2] = figure[:,:,1]

fig,axs = plt.subplots(1, 2, layout='constrained', figsize = [13,13])

axs[0].imshow(figure)

axs[0].set_title('Imagem original / Original Image')

axs[0].set_axis_off()

axs[1].imshow(figure_cp)

axs[1].set_title('Imagem modificada / Modified Image')

axs[1].set_axis_off()

plt.show()
```

```python
import numpy as np
from google.colab.patches import cv2_imshow

# Sétima modificação efetuada na segunda imagem
# Seventh modification made to the second image

# Mesmos passos da sétima modificação realizada na primeira imagem
# Same steps as the seventh modification performed on the first image.

figure = cv2.cvtColor(figure, cv2.COLOR_RGB2BGR)

high_booster = np.array([
    [-1,-1,-1],
    [-1,8,-1],
    [-1,-1,-1]
])

figure_filtered = cv2.filter2D(figure, -1, high_booster)

cv2_imshow(figure_filtered)
```

```python


# Oitava modificação efetuada na segunda imagem
# Eighth modification made to the second image

# Mesmos passos da oitava modificação realizada na primeira imagem
# Same steps as the eighth modification performed on the first image.

figure_gray = cv2.cvtColor(figure, cv2.COLOR_BGR2GRAY)

high_booster = np.array([
    [-1,-1,-1],
    [-1,8,-1],
    [-1,-1,-1]
])

figure_filtered = cv2.filter2D(figure_gray, -1, high_booster)

cv2_imshow(figure_filtered)

```

```python


# Nona modificação efetuada na imagem
# Ninth modification made to the image

# Mesmos passos da nona modificação realizada na primeira imagem
# Same steps as the ninth modification performed on the first image.

height = figure_filtered.shape[0]

width = figure_filtered.shape[1]

figure_filtered_cp = figure_filtered.copy()

for i in range(height):
  for j in range(width):
    if figure_filtered[i,j]>100:
      figure_filtered_cp[i,j]=255


cv2_imshow(figure_filtered_cp)

```

```python


# Décima modificação efetuada na imagem
# Tenth modification made to the image

# Mesmos passos da décima modificação realizada na primeira imagem
# Same steps as the tenth modification performed on the first image.

height = figure_filtered.shape[0]
width = figure_filtered.shape[1]

figure_filtered_cp = figure_filtered.copy()

for i in range(height):
  for j in range(width):
    if figure_filtered[i,j]>50:
      figure_filtered_cp[i,j]=255


cv2_imshow(figure_filtered_cp)
```

```python

# Décima primeira modificação efetuada na imagem
# Eleventh modification made to the image

# Mesmos passos da décima primeira modificação realizada na primeira imagem
# Same steps as the eleventh modification performed on the first image.

height = figure_filtered.shape[0]

width = figure_filtered.shape[1]

figure_filtered_cp = figure_filtered.copy()

for i in range(height):
  for j in range(width):
    if figure_filtered[i,j]>100:
      figure_filtered_cp[i,j]=255

dim = 5

c=1/9

mask = np.ones((dim,dim))*c

figure_filtered_cp = cv2.filter2D(figure_filtered_cp, -1, mask)

cv2_imshow(figure_filtered_cp)
```

```python




# Décima segunda modificação efetuada na imagem
# Twelfth modification made to the image

# Mesmos passos da décima segunda modificação realizada na primeira imagem, mas agora com o limiar de intensidade igual a 20
# Same steps as the twelfth modification performed on the first image, but now with the intensity threshold equal to 20

height = figure_filtered.shape[0]
width = figure_filtered.shape[1]

figure_filtered_cp = figure_filtered.copy()

for i in range(height):
  for j in range(width):
    if figure_filtered[i,j]>20:
      figure_filtered_cp[i,j]=255

dim = 5

c=1/9

mask = np.ones((dim,dim))*c

figure_filtered_cp = cv2.filter2D(figure_filtered_cp, -1, mask)

cv2_imshow(figure_filtered_cp)
```
