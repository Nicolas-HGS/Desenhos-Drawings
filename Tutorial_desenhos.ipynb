```python
# @title
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
import numpy as np
import time
from google.colab import output

def total_func():

  print("Digite o nome do arquivo da imagem")
  img_flname = str(input())
  c=0

  while c==0:
    try:
      figure=cv2.imread(img_flname)
      cv2_imshow(figure)
      time.sleep(1)
      c=1
    except:
      print("Nome do arquivo digitado não existe, por favor digitar novamente:")
      img_flname = str(input())

  height = figure.shape[0]

  width = figure.shape[1]

  print("Digite qual operação deseja realizar. As opções são: troca de cores, obter contornos e realçar contornos.")
  operation = str(input())
  control=0
  p=0

  while operation != "finalizar":

    def visu(img_1,img_2):

      print("Deseja visualizar lado a lado a imagem atual e a original? Responda escrevendo sim ou não")
      print("A depender das imagens, o tempo de espera pode ser superior a 10 segundos. Aguarde as imagens aparecerem antes de continuar")
      answer = str(input())
      if answer=="sim":
        cv2_imshow(cv2.hconcat((img_1,img_2)))

    def salv(fig):
      print("Deseja salvar a imagem? Responda escrevendo sim ou não")
      answer = str(input())
      if answer=="sim":
        print("Digite o nome que deseja salvar a imagem junto da extensão do arquivo")
        print("Se por acaso a imagem não aparecer após o salvamento, atualize a página")
        nome_arquivo = str(input())
        cv2.imwrite(nome_arquivo, fig)

    def limp():
      print("Deseja limpar a saída? Responda escrevendo sim ou não")
      answer = str(input())
      if answer=="sim":
        print("Digite qual operação deseja realizar. As opções são: troca de cores, obter contornos e realçar contornos.")
        print("Do contrário, digite finalizar")
        operation = str(input())
        output.clear()
        time.sleep(1)
        return(operation)
      else:
        print("Digite qual operação deseja realizar. As opções são: troca de cores, obter contornos e realçar contornos.")
        print("Do contrário, digite finalizar")
        operation = str(input())
        return(operation)

    if operation == "troca de cores":
      control=1
      figure_cp = figure.copy()
      print("Indique os pixels de uma determinada camada de cor com os quais você quer igualar os pixels da camada azul: 0 - Azul, 1 - Verde, 2 - Vermelho")
      camada = int(input())
      figure_cp[:,:,0] = figure[:,:,camada]
      print("Indique os pixels de uma determinada camada de cor com os quais você quer igualar os pixels da camada verde: 0 - Azul, 1 - Verde, 2 - Vermelho")
      camada = int(input())
      figure_cp[:,:,1] = figure[:,:,camada]
      print("Indique os pixels de uma determinada camada de cor com os quais você quer igualar os pixels da camada vermelha: 0 - Azul, 1 - Verde, 2 - Vermelho")
      camada = int(input())
      figure_cp[:,:,2] = figure[:,:,camada]
      cv2_imshow(figure_cp)
      time.sleep(1)
      visu(figure,figure_cp)
      salv(figure_cp)
      operation=limp()

    elif operation == "obter contornos":
      figure_gray = cv2.cvtColor(figure, cv2.COLOR_BGR2GRAY)
      high_booster = np.array([
                      [-1,-1,-1],
                      [-1,8,-1],
                      [-1,-1,-1]
                      ])

      figure_filtered = cv2.filter2D(figure_gray, -1, high_booster)
      figure_filtered_3layers = figure.copy()
      figure_filtered_3layers[:,:,0] = figure_filtered
      figure_filtered_3layers[:,:,1] = figure_filtered
      figure_filtered_3layers[:,:,2] = figure_filtered
      cv2_imshow(figure_filtered)
      time.sleep(1)
      visu(figure,figure_filtered_3layers)
      salv(figure_filtered_3layers)
      operation=limp()

    elif operation == "realçar contornos":
      figure_cp = figure.copy()
      print("digite um valor para o parâmetro de realce (0-255):")
      d = int(input())
      for i in range(height):
        for j in range(width):
          if figure_cp[i,j,0]>=d:
            figure_cp[i,j,0]=255
      figure_cp[:,:,1]=figure_cp[:,:,0]
      figure_cp[:,:,2]=figure_cp[:,:,0]
      cv2_imshow(figure_cp)
      time.sleep(1)
      visu(figure,figure_cp)
      salv(figure_cp)
      operation=limp()

    else:
      if operation != "finalizar" and operation != "troca de cores" and operation != "obter contornos" and operation != "realçar contornos":
        print("Não existe operação com esse nome, ou ela foi digitada incorretamente. Por favor, digite novamente.")
        print("As opções são: troca de cores, obter contornos e realçar contornos.")
        print("Caso queira finalizar o código, digite finalizar")


        operation = str(input())

total_func()
```
